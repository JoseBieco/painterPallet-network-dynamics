## Passo 1: Configuração e Leitura de Dados
**Descrição:** Importação das bibliotecas essenciais (`pandas`, `networkx`, `matplotlib`, `powerlaw`) e carregamento da base de dados bruta do MoMA (`Artworks.csv`). O objetivo aqui é garantir que o ambiente está pronto e os dados estão acessíveis.

## Passo 2: Pré-processamento e Limpeza (Data Wrangling)
**Descrição:** Filtragem do dataset para isolar apenas as obras classificadas como "Painting" (Pintura). Limpeza da coluna de datas utilizando expressões regulares para extrair o ano exato de produção, removendo registros nulos ou inválidos.

## Passo 3: Construção do Grafo Bipartido Temporal
**Descrição:** Modelagem inicial dos dados em uma rede bipartida contendo dois conjuntos de nós: **Pintores** e **Anos**. A criação das arestas se dá conectando um pintor a um ano específico em que ele produziu uma ou mais obras, estabelecendo a base temporal da nossa rede.

## Passo 4: Projeção Ponderada e Extração da Componente Principal
**Descrição:** Transformação do grafo bipartido em uma rede unipartida de pintores (`Pintor -> Pintor`). O peso da aresta será equivalente ao número de anos de produção simultânea. Aplicação de um *threshold* (filtro de peso mínimo) para remover conexões fracas e extração da Maior Componente Conexa (*Giant Component*) para garantir que a rede não tenha ilhas isoladas que inviabilizem as simulações dinâmicas.

## Passo 5: Extração de Métricas Topológicas e Centralidades
**Descrição:** Cálculo das métricas globais da rede (número de nós, arestas, densidade, grau médio e coeficiente de aglomeração). Extração de métricas locais vitais para entender o papel dos artistas, criando um ranking com *Degree Centrality* (popularidade), *Betweenness Centrality* (pontes entre movimentos) e *Eigenvector Centrality* (influência na elite).

## Passo 6: Análise da Distribuição de Grau (Scale-Free)
**Descrição:** Plotagem da distribuição de probabilidade do grau $P(k)$ em escalas linear e Log-Log. Utilização do pacote `powerlaw` para ajustar os dados e testar matematicamente se a rede de pintores segue uma distribuição de Lei de Potência (comportamento de cauda longa indicando a presença de *hubs*).

## Passo 7: Visualização da Topologia Estática
**Descrição:** Plotagem gráfica da rede utilizando `matplotlib` e `networkx` (ou `pyvis`). O tamanho dos nós será proporcional ao grau (ou centralidade) do artista, evidenciando visualmente a estrutura de núcleo-periferia e os maiores expoentes da pintura moderna.

## Passo 8: Processos Dinâmicos - Simulação Epidêmica (Modelo SIR)
**Descrição:** Implementação do modelo `Suscetível-Infectado-Recuperado` sobre a rede estática. O "vírus" simula a propagação de uma nova técnica ou estilo artístico.
- **Cenário A:** O paciente zero é o maior hub da rede (o artista com maior centralidade).
- **Cenário B:** O paciente zero é um nó periférico (artista pouco conectado).
- Plotagem das curvas `S(t)`, `I(t)` e `R(t)` ao longo do tempo (iterações) para comparar a velocidade de "contágio" da ideia na rede.

## Passo 9: Discussão dos Resultados e Conclusão
**Descrição:** Interpretação das simulações dinâmicas à luz das métricas topológicas. Como a presença de *hubs* (identificados no Passo 6) acelerou a difusão da técnica no Passo 8? Considerações finais sobre o uso de redes complexas para entender a evolução quantitativa da História da Arte.
